```python
def initializeDatabase():
    import pandas as pd
    df=pd.read_csv('database.csv')
    return df
```

```python
def personalinfo():
    name= input("Enter your name")
    import re
    while True:
        phone= input("Enter your phone no.")
        if phone.isdigit() and len(phone) == 10:
            break
        else: print ('number not valid')
    while True:
        dob=input("Enter your date of birth [DD/MM/YYYY]")
        pattern = r'^\d{2}/\d{2}/\d{4}$'
        if re.match(pattern, dob):
            break
        else:
            print("Invalid date format. Please use DD/MM/YYYY format.") 
        
    return name,phone,dob

#personalinfo()
```

```python
def takemail():
    import re
    while True:
        mail = input("Enter your email address: ").lower()
        pattern = r'^[a-z0-9._%+-]+@[a-z0-9.-]+\.com$'
        
        if re.match(pattern, mail):
            return mail
        else:
            print("Invalid email format. Please try again.")

#takemail()
```

```python
def password():
    import re
    while True:
        passw= input("enter your password")
        if len(passw) < 8:
            print ("Password is too short (min 8 characters).")
        if not re.search(r"[A-Z][a-z][0-9]", passw):
            print ("Password needs at least one uppercase and lowercase letter and a number.")
        if not re.search(r"[!@#$%^&*()_+\-=\[\]{}|\\;:'\",.<>/?]", passw):
            print ("Password needs at least one special character.")
        else:
            break
    return passw

#password()
```
